```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = r"C:\Users\Admin\Desktop\B23AI065_CVIP\mouse.jpg"
```

```python
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE) # Load image as grayscale (0-255 pixel values)
```

```python
# Step 1a: Check if image is loaded successfully 
if image is None:
    print("Error: Image not found or unable to load!") # Show error if image path is wrong 
else: 
    print("Image loaded successfully!") # Confirm image loaded
```

```python
# Step 2: Apply basic intensity transformation functions 
#  2.1. Negative Transformation 
negative_image = 255 - image # Invert pixel values: dark becomes bright, bright becomes dark 
print("Negative transformation applied.") # Confirm transformation done
```

```python
# 2.2. Logarithmic Transformation
c = 255 / np.log(1 + np.max(image)) # Scale constant to map log values back to 0-255
log_image = c * (np.log(1 + image.astype(np.float32))) # Apply log transformation, convert image to float for log
log_image = np.uint8(log_image) # Convert float result back to unsigned 8-bit integers (0-255)
print("Logarithmic transformation applied.") # Confirm transformation done
```

```python
# 2.3. Power-Law (Gamma) Transformation
gamma = 2.2 # Gamma value; >1 darkens image, <1 brightens image
gamma_corrected = np.array(255 * ((image / 255) ** gamma), dtype='uint8') # Apply gamma correction
print("Power-law (gamma) transformation applied.") # Confirm transformation done
```

```python
# Step 3: Display original and transformed images
titles = ['Original Image', 'Negative Image', 'Logarithmic Transformation', 'Gamma Transformation'] # Titles for display
images = [image, negative_image, log_image, gamma_corrected] # List of images to display
for i in range(4): # Loop to display all images
    plt.subplot(2, 2, i + 1) # Create 2x2 grid for subplots 
    plt.imshow(images[i], cmap='gray') # Show image in grayscale 
    plt.title(titles[i]) # Set subplot title 
    plt.axis('off') # Hide axes for cleaner display
plt.tight_layout() # Adjust spacing between subplots 
plt.show() # Show all images 
print("Transformations displayed successfully.") # Confirm display done
```
