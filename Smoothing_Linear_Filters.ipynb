```python
import cv2                     # For image processing
import numpy as np             # For numerical operations
import matplotlib.pyplot as plt  # For displaying images
```

```python
# Function to read and process a real image

def process_real_image(image_path):
    """
    Read and apply smoothing filters to a real image.
    """

    # Step 1: Read the input image in grayscale
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Step 2: Apply smoothing linear filters
    def apply_smoothing_filters(image):
        """
        Apply box and Gaussian filters to the input image.
        """
        box_filter = cv2.blur(image, (5, 5))             # 5x5 averaging filter
        gaussian_filter = cv2.GaussianBlur(image, (5, 5), 0)  # 5x5 Gaussian filter
        return box_filter, gaussian_filter

    # Step 3: Display results
    def display_results(original, box, gaussian, title_prefix="Real Image"):
        """
        Display the original and filtered images.
        """
        plt.figure(figsize=(12, 8))

        plt.subplot(1, 3, 1)
        plt.title(f"{title_prefix} - Original")
        plt.imshow(original, cmap='gray')
        plt.axis('off')

        plt.subplot(1, 3, 2)
        plt.title(f"{title_prefix} - Box Filter")
        plt.imshow(box, cmap='gray')
        plt.axis('off')

        plt.subplot(1, 3, 3)
        plt.title(f"{title_prefix} - Gaussian Filter")
        plt.imshow(gaussian, cmap='gray')
        plt.axis('off')

        plt.tight_layout()
        plt.show()

    box_result, gaussian_result = apply_smoothing_filters(image)
    display_results(image, box_result, gaussian_result)
```

```python

# Function to create and process a synthetic noisy image

def process_synthetic_image():
    """
    Create a synthetic noisy matrix and apply smoothing filters.
    """

    # Step 1: Create a synthetic noisy matrix
    def create_noisy_matrix():
        matrix = np.zeros((9, 9), dtype=np.uint8)   # 9x9 black matrix
        matrix[2:7, 2:7] = 150                      # Gray square in the center
        noise = np.random.randint(0, 50, (9, 9), dtype=np.uint8)
        noisy_matrix = cv2.add(matrix, noise)
        return noisy_matrix

    noisy_matrix = create_noisy_matrix()

    # Step 2: Apply smoothing filters
    def apply_smoothing_filters(matrix):
        box_filter = cv2.blur(matrix, (3, 3))              # 3x3 box filter
        gaussian_filter = cv2.GaussianBlur(matrix, (3, 3), 0)  # 3x3 Gaussian filter
        return box_filter, gaussian_filter

    box_result, gaussian_result = apply_smoothing_filters(noisy_matrix)

    # Step 3: Display results
    def display_results(original, box, gaussian, title_prefix="Synthetic Image"):
        plt.figure(figsize=(12, 6))

        plt.subplot(1, 3, 1)
        plt.title(f"{title_prefix} - Original with Noise")
        plt.imshow(original, cmap='gray')
        plt.colorbar()
        plt.axis('off')

        plt.subplot(1, 3, 2)
        plt.title(f"{title_prefix} - Box Filter")
        plt.imshow(box, cmap='gray')
        plt.colorbar()
        plt.axis('off')

        plt.subplot(1, 3, 3)
        plt.title(f"{title_prefix} - Gaussian Filter")
        plt.imshow(gaussian, cmap='gray')
        plt.colorbar()
        plt.axis('off')

        plt.tight_layout()
        plt.show()

    display_results(noisy_matrix, box_result, gaussian_result)

```
