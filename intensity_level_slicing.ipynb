```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = r'C:\Users\Admin\Desktop\B23AI065_CVIP\mouse.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
# Step 1a: Check if image is loaded successfully 
if image is None:
    print("Error: Image not found or unable to load!") # Show error if image path is wrong 
else: 
    print("Image loaded successfully!") # Confirm image loaded
```

```python
def binary_slicing(img, r_min, r_max):
    out = np.zeros_like(img, dtype=np.uint8)
    mask = (img >= r_min) & (img <= r_max)
    out[mask] = 255
    return out

def gray_level_slicing(img, r_min, r_max, new_value=255):
    out = img.copy()
    mask = (img >= r_min) & (img <= r_max)
    out[mask] = new_value
    return out
```

```python
r_min, r_max = 180, 255

binary_result = binary_slicing(image, r_min, r_max)
gray_result = gray_level_slicing(image, r_min, r_max)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(binary_result, cmap='gray')
plt.title('Binary Slicing (Fig. 3.11(a))')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(gray_result, cmap='gray')
plt.title('Gray-Level Slicing (Fig. 3.11(b))')
plt.axis('off')

plt.tight_layout()
plt.show()
```
